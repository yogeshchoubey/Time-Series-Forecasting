```python
import os
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
filename = 'Daily Bike Sharing.csv'
path = '..\Dataset'
filepath = os.path.join(path,filename)
#print(filepath)
```

```python
df = pd.read_csv(filepath, 
                 index_col='dteday',
                 parse_dates=True)
```

```python
df.index
```

```python
df.info()
```

```python
df.describe()
```

```python
df['cnt'].plot()
```

```python
result = seasonal_decompose(df['cnt'],model='multiplicative')
```

```python
result.plot()
```

```python
result.trend.plot();
```

```python
result.seasonal.plot();
```

```python
result.resid.plot();
```
